```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Make sample of random points 

x,y = make_regression(n_samples=200,n_features=1,noise=25.5,n_informative=1,random_state=1)
```

```python
#Plotting these points
plt.scatter(x,y)
plt.show()
```

```python
#Making the object of Linear Regression 
lr = LinearRegression()
# Fitting our data in the object
lr.fit(x,y)
```

```python
# Taking 10 test point and plotting the line of equation using these point and the values of theta 0 and theta 1

#Here linspace() function will return 10 equally spaced points between -3 and 3

x_test = np.linspace(-3,3,10)
# predicting the values of y using inbuilt linear regression model
y_pred = lr.predict(x_test.reshape(-1,1))

```

```python
# Plotting our traning data and predicted line in the same figure. This red line is our final result.

plt.scatter(x,y)
plt.plot(x_test,y_pred,color = "red")
plt.show()
```

```python
# Accuracy of the model
lr.score(x,y)
```
